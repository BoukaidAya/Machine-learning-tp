1. `CLIENTNUM`: Il s'agit d'un identifiant unique attribué à chaque client dans l'ensemble de données. C'est un numéro de client ou d'utilisateur qui sert à distinguer les différents clients.

2. `Attrition_Flag`: Cette colonne indique si le client est en attrition ou non. L'attrition se réfère généralement à la perte de clients ou à leur désengagement d'une entreprise.

3. `Customer_Age`: C'est l'âge du client.

4. `Gender`: C'est le genre du client, généralement représenté par des catégories telles que "Male" (masculin) ou "Female" (féminin).

5. `Dependent_count`: Il s'agit du nombre de personnes à charge du client, comme expliqué précédemment.

6. `Education_Level`: C'est le niveau d'éducation du client, par exemple, "Graduate" (diplômé), "High School" (lycée), etc.

7. `Marital_Status`: C'est l'état civil du client, comme "Married" (marié), "Single" (célibataire), etc.

8. `Income_Category`: C'est la catégorie de revenu du client, souvent répartie en tranches comme "Less than $40K" (moins de 40 000 $), "$40K - $60K" (40 000 $ - 60 000 $), etc.

9. `Card_Category`: Il s'agit de la catégorie de carte de crédit détenue par le client, comme "Blue", "Silver", "Gold", "Platinum", etc.

10. `Months_on_book`: C'est le nombre de mois pendant lesquels le client a été titulaire de son compte.

11. `Total_Relationship_Count`: Il s'agit du nombre total de produits financiers que le client détient auprès de la banque ou de l'institution financière.

12. `Months_Inactive_12_mon`: C'est le nombre de mois au cours des 12 derniers mois pendant lesquels le client n'a pas effectué de transaction.

13. `Contacts_Count_12_mon`: Il s'agit du nombre de contacts que le client a eu avec la banque ou l'institution financière au cours des 12 derniers mois.

14. `Credit_Limit`: C'est la limite de crédit du client, c'est-à-dire le montant maximum qu'il peut emprunter sur sa carte de crédit ou son compte.

15. `Total_Revolving_Bal`: Comme expliqué précédemment, c'est le solde total des crédits renouvelables sur le compte du client.

16. `Avg_Open_To_Buy`: C'est la moyenne du montant disponible pour l'emprunt sur la carte de crédit ou le compte.

17. `Total_Amt_Chng_Q4_Q1`: C'est le changement total du montant dépensé entre le quatrième et le premier trimestre.

18. `Total_Trans_Amt`: C'est le montant total des transactions effectuées par le client.

19. `Total_Trans_Ct`: C'est le nombre total de transactions effectuées par le client.

20. `Total_Ct_Chng_Q4_Q1`: C'est le changement total dans le nombre de transactions entre le quatrième et le premier trimestre.

21. `Avg_Utilization_Ratio`: C'est le rapport moyen entre le solde utilisé et la limite de crédit disponible.
